```python
import pandas as pd
```

```python
movies = pd.read_csv(
    "/Users/apple/Downloads/movie.csv",
    header=None,
    names=["MovieID", "Year", "Title"],
    encoding="latin-1",
    engine="c",
    on_bad_lines="skip"
)
```

```python

movies.info()
```

```python
movies.head()
```

```python
movies["MovieID"] = movies["MovieID"].astype("int32")
movies["Year"] = pd.to_numeric(movies["Year"], errors="coerce")
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.to_csv("cleaned_movies.csv", index=False)
```
