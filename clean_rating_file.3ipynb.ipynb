```python
import pandas as pd

raw = pd.read_csv(
    "/Users/apple/Downloads/combined_data_1.txt",
    header=None,
    names=["raw"],
    engine="c",
    sep="|"
)
```

```python
raw["MovieID"] = raw["raw"].str.extract(r"(\d+):")
raw["MovieID"] = raw["MovieID"].ffill()

ratings = raw[~raw["raw"].str.contains(":")]

ratings[["UserID","Rating","Date"]] = ratings["raw"].str.split(",", expand=True)

df = ratings[["MovieID","UserID","Rating","Date"]]

df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

```python
df["MovieID"] = df["MovieID"].astype("int32")
df["UserID"] = df["UserID"].astype("int32")
df["Rating"] = df["Rating"].astype("float32")
df["Date"] = pd.to_datetime(df["Date"])

df.reset_index(drop=True, inplace=True)
```
